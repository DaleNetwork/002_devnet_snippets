```python
import requests
import json

requests.packages.urllib3.disable_warnings()

def gettoken():
    url = 'https://sandboxdnac.cisco.com/dna/system/api/v1/auth/token'
    USER = 'devnetuser'
    PASS = 'Cisco123!'
    response = requests.request('POST', url, auth = (USER, PASS))
    return response.json()["Token"]

token = gettoken()
# print(token)

url = 'https://sandboxdnac.cisco.com/dna/intent/api/v1/network-device/config'
payload = None

headers = {
    "Content-Type": "application/json",
    "Accept": "application/json",
    "X-Auth-token": token
}

response = requests.request('GET',url,headers=headers, data = payload)
print(response)
#print(response.text)

# print(response.json()) # without .json(), response only has 
parsed_response = json.loads(response.text.encode('utf8'))
print('The type of parsed_response is {}, its length is {}'.format(type(parsed_response["response"]), len(parsed_response["response"])))
parsed1 = response.text.encode('utf8')
parsed2 = response.json()
print ('Type("response.text") is {}, Type("response.json()") is {}'.format(type(parsed1),type(parsed2)))
pretty_response = json.dumps(parsed_response, indent = 2)
#print(pretty_response)
print(parsed_response["response"][2]["runningConfig"])
```
