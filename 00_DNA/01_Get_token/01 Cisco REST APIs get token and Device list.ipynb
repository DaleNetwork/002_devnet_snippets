```python
import requests   # quick start, https://docs.python-requests.org/en/latest/user/quickstart/#response-content
import json  # print pretty json

# Use self signed certs
requests.packages.urllib3.disable_warnings()

#Credentials
USER = 'devnetuser'
PASS = 'Cisco123!'

# URL for POST request for DNA Center, to get token
url1 = 'https://sandboxdnac.cisco.com/dna/system/api/v1/auth/token'

#Run POST
response = requests.post(url1,auth=(USER,PASS))

#print result
print(response)
print('Status Code:' + str(response.status_code))
print('Response Text: ' + response.text+'\n')            # this is the token

token=response.json()['Token']         #json() method
print("token = ",token)

# print(response.json())

# RUN GET,  device list

payload = {}

headers = {
    "X-Auth-Token": token,
    "Contert-Type": "application/json",
    "Accept": "application/json"
}
url2 = 'https://sandboxdnac.cisco.com/dna/intent/api/v1/network-device'
response = requests.request('GET', url2, headers=headers, data=payload)
#print(response.json())
parsedlist= json.loads(response.text)
#print(parsedlist)
print(json.dumps(parsedlist, indent = 2, sort_keys = False))
```
